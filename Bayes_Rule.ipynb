```python
# calculate P(A|B) given P(A), P(B|A), P(B|not A)
def bayes_theorem( p_a
                  , p_b_given_a
                  , p_b_given_not_a):
    
	# calculate P(not A)
    not_a = 1 - p_a
    print('P(not A) = 1-P(A) : {}'.format(not_a))
    print()
    
	# calculate P(B)
    p_b = p_b_given_a * p_a + p_b_given_not_a * not_a
    print('P(B) = P(B|A)*P(A) + P(B|not A)*P(not A) : {}'.format(not_a))
    print()
    
	# calculate P(A|B)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    print("P(A|B) = (P(B|A)*P(A))/P(B) :{}".format(p_a_given_b))
    print()
    
    return p_a_given_b
 
# P(A)
p_a = 0.0002
print('P(A) : {}'.format(p_a))
print()

# P(B|A)
p_b_given_a = 0.85
print('P(B|A) : {}'.format(p_b_given_a))
print()

# P(B|not A)
p_b_given_not_a = 0.05
print('P(B|not A) : {}'.format(p_b_given_not_a))
print()

# calculate P(A|B)
result = bayes_theorem( p_a
                       , p_b_given_a
                       , p_b_given_not_a )

# summarize
print('P(A|B) = %.3f%%' % (result * 100))
print()
```
